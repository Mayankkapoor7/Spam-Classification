```python
import os
import numpy as np
import pandas as pd
dataset=pd.read_csv('/content/SPAM_text_message.csv')
```

```python
dataset.head()
```

```python
dataset['Category'].value_counts()/len(dataset)
```

```python
dataset.isnull().sum()
```

```python
ham=dataset[dataset['Category']=='ham']
spam=dataset[dataset['Category']=='spam']
```

```python
print(ham.shape, spam.shape)
```

```python
ham
```

```python
spam
```

```python
# Balance the dataset
spam=spam.sample(ham.shape[0],replace=True)# replace=True will allow us to duplicate the spam data and make it equal to ham.shape[0]
```

```python
ham.shape[0]
```

```python
spam.shape[0]
```

```python
data=pd.concat([ham,spam])
```

```python
data
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'data' is a DataFrame with 'Message' and 'Category' columns
x_train, x_test, y_train, y_test = train_test_split(
    data['Message'],  # Features (independent variable)
    data['Category'],  # Labels (dependent variable)
    test_size=0.3,  # 30% of the data will be used as the test set
    random_state=42  # Ensures reproducibility of the split
)
```

# Feature Extraction and Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
classifier = Pipeline([('bow', CountVectorizer()),
                       ('rfmodel', RandomForestClassifier())])
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred_train=classifier.predict(x_train)
y_pred_test=classifier.predict(x_test)
```

```python
y_pred_test
```

```python
# Evaluate the model
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_train,y_pred_train))
print("*****************************************")
print(confusion_matrix(y_test,y_pred_test))
```

```python
print(classification_report(y_train,y_pred_train))
print("*****************************************")
print(classification_report(y_test,y_pred_test))
```

```python
print("Training accuracy : ", accuracy_score(y_train,y_pred_train))
print("*****************************************")
print("Test accuracy : ",accuracy_score(y_test,y_pred_test))
```
